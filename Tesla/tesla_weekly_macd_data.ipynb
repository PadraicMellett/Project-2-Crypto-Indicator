```python
import pandas as pd
import requests 
```

```python
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY&symbol=TSLA&apikey=PLWHFCZSI5GFFUMN'
r = requests.get(url)
tesla_weekly_data = r.json()

print(tesla_weekly_data)
```

```python
tesla_weekly_df = pd.DataFrame.from_dict(tesla_weekly_data['Weekly Time Series']).T
tesla_weekly_df.head(50)
```

```python
url = 'https://www.alphavantage.co/query?function=MACD&symbol=TSLA&interval=weekly&series_type=open&apikey=PLWHFCZSI5GFFUMN'
r = requests.get(url)
tesla_macd_weekly = r.json()

print(tesla_macd_weekly)
```

```python
tesla_macd_weekly_df = pd.DataFrame.from_dict(tesla_macd_weekly['Technical Analysis: MACD']).T
tesla_macd_weekly_df.head(50)
```

```python
tesla_weekly_df = tesla_weekly_df.reset_index()
```

```python
tesla_macd_weekly_df = tesla_macd_weekly_df.reset_index()
```

```python
merged_weekly_macd = pd.merge(tesla_weekly_df, tesla_macd_weekly_df , on=['index', 'index'])
merged_weekly_macd.head(50)
```

```python
merged_weekly_macd = merged_weekly_macd.rename(columns={'index': 'Date', '1. open': 'Open', '2. high': 'High', '3. low': 'Low', '4. close': 'Close', '5. volume': 'Volume', 'MACD_Signal': 'MACD', 'MACD_Hist': 'Hist', 'MACD': 'MACD Value'})

merged_weekly_macd
```

```python
merged_weekly_macd.dropna()
```

```python
merged_weekly_macd_adjusted = merged_weekly_macd[['Date', 'Close', 'Hist']]
merged_weekly_macd_adjusted
```

```python
merged_weekly_macd_adjusted = merged_weekly_macd_adjusted.sort_values('Date', ascending = True)
merged_weekly_macd_adjusted.head()
```

```python
merged_weekly_macd_adjusted.dtypes
```

```python
merged_weekly_macd_adjusted['Close'] = pd.to_numeric(merged_weekly_macd_adjusted['Close'], errors='coerce')
```

```python
merged_weekly_macd_adjusted['Hist'] = pd.to_numeric(merged_weekly_macd_adjusted['Hist'], errors='coerce')
```

```python
merged_weekly_macd_adjusted.dtypes
```

```python
merged_weekly_macd_adjusted['Buy'] = (merged_weekly_macd_adjusted['Hist'] > 0) & (merged_weekly_macd_adjusted['Hist'].shift(periods = 1) < 0)
```

```python
merged_weekly_macd_adjusted['Sell'] = (merged_weekly_macd_adjusted['Hist'] < 0) & (merged_weekly_macd_adjusted['Hist'].shift(periods = 1) > 0)
```

```python
merged_weekly_macd_adjusted.head(50)
```

```python
df = merged_weekly_macd_adjusted[(merged_weekly_macd_adjusted['Sell'] == True) | (merged_weekly_macd_adjusted['Buy'] == True)] 
df = df.reset_index()
df
```

```python
import numpy as np
```

```python
cap = 10000
shares = 0
cap_list = []
share_list = []
df_test = df.copy()

for i, row in df_test.iterrows():
    if row['Sell']:
        #print("sell")
        cap = shares * row['Close']
        shares = 0
        #print(f"1 cap = {cap} \t shares={shares}")

        cap_list.append(cap)
        share_list.append(0)
    else:
        #print("buy")
        shares = cap / row['Close']
        cap = 0
        #print(f"2 cap = {cap} \t shares={shares}")

        cap_list.append(0)
        share_list.append(shares)

    if cap < 0:
        print("BROKE!!!")
        break

if cap >= 0:
    df_test['capital'] = cap_list
    df_test['shares'] = share_list

df_test
```
