```python
import pandas as pd
import requests 
```

```python
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY&symbol=TSLA&apikey=PLWHFCZSI5GFFUMN'
r = requests.get(url)
tesla_weekly_data = r.json()

print(tesla_weekly_data)
```

```python
tesla_weekly_df = pd.DataFrame.from_dict(tesla_weekly_data['Weekly Time Series']).T
tesla_weekly_df.head(50)
```

```python
url = 'https://www.alphavantage.co/query?function=STOCH&symbol=TSLA&interval=daily&time_period=14&apikey=PLWHFCZSI5GFFUMN'
r = requests.get(url)
tesla_stoch_weekly = r.json()

print(tesla_stoch_weekly)
```

```python
tesla_stoch_weekly_df = pd.DataFrame.from_dict(tesla_stoch_weekly['Technical Analysis: STOCH']).T
tesla_stoch_weekly_df.head(50)
```

```python
tesla_weekly_df = tesla_weekly_df.reset_index()
```

```python
tesla_stoch_weekly_df = tesla_stoch_weekly_df.reset_index()
```

```python
merged_weekly_stoch = pd.merge(tesla_weekly_df, tesla_stoch_weekly_df , on=['index', 'index'])
merged_weekly_stoch.head(50)
```

```python
merged_weekly_stoch = merged_weekly_stoch.rename(columns={'index': 'Date', '1. open': 'Open', '2. high': 'High', '3. low': 'Low', '4. close': 'Close', '5. volume': 'Volume'})

merged_weekly_stoch
```

```python
merged_weekly_stoch.dropna()
```

```python
merged_weekly_stoch_adjusted = merged_weekly_stoch[['Date', 'Close', 'SlowK']]
merged_weekly_stoch_adjusted
```

```python
merged_weekly_stoch_adjusted = merged_weekly_stoch_adjusted.sort_values('Date', ascending = True)
merged_weekly_stoch_adjusted.head()
```

```python
merged_weekly_stoch_adjusted.dtypes
```

```python
merged_weekly_stoch_adjusted['Close'] = pd.to_numeric(merged_weekly_stoch_adjusted['Close'], errors='coerce')
```

```python
merged_weekly_stoch_adjusted['SlowK'] = pd.to_numeric(merged_weekly_stoch_adjusted['SlowK'], errors='coerce')
```

```python
merged_weekly_stoch_adjusted.dtypes
```

```python
merged_weekly_stoch_adjusted['Buy'] = (merged_weekly_stoch_adjusted['SlowK'] < 20) & (merged_weekly_stoch_adjusted['SlowK'].shift(periods = 1) > 20)
```

```python
merged_weekly_stoch_adjusted['Sell'] = (merged_weekly_stoch_adjusted['SlowK'] > 80) & (merged_weekly_stoch_adjusted['SlowK'].shift(periods = 1) < 80)
```

```python
merged_weekly_stoch_adjusted.head(50)
```

```python
df = merged_weekly_stoch_adjusted[((merged_weekly_stoch_adjusted['Sell'] == True) & (merged_weekly_stoch_adjusted['Sell'].shift(periods = 1) == False)) | ((merged_weekly_stoch_adjusted['Buy'] == True) & (merged_weekly_stoch_adjusted['Buy'].shift(periods = 1) == False))] 
df = df.reset_index()
df
```

```python
df2 = df[((df['Sell'] == True) & (df['Sell'].shift(periods = 1) == False)) | ((df['Buy'] == True) & (df['Buy'].shift(periods = 1) == False))] 
df2 = df2.reset_index()
df2.head(50)
```

```python
import numpy as np
```

```python
cap = 10000
shares = 0
cap_list = []
share_list = []
df_test = df2.copy()

for i, row in df_test.iterrows():
    if row['Buy']:
        #print("sell")
        cap = shares * row['Close']
        shares = 0
        #print(f"1 cap = {cap} \t shares={shares}")

        cap_list.append(cap)
        share_list.append(0)
    else:
        #print("buy")
        shares = cap / row['Close']
        cap = 0
        #print(f"2 cap = {cap} \t shares={shares}")

        cap_list.append(0)
        share_list.append(shares)

    if cap < 0:
        print("BROKE!!!")
        break

if cap >= 0:
    df_test['capital'] = cap_list
    df_test['shares'] = share_list

df_test
```
