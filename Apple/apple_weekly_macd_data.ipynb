```python
import pandas as pd
import requests 
```

```python
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY&symbol=AAPL&apikey=PLWHFCZSI5GFFUMN'
r = requests.get(url)
apple_weekly_data = r.json()

print(apple_weekly_data)
```

```python
apple_weekly_df = pd.DataFrame.from_dict(apple_weekly_data['Weekly Time Series']).T
apple_weekly_df.head(50)
```

```python
url = 'https://www.alphavantage.co/query?function=MACD&symbol=AAPL&interval=weekly&series_type=open&apikey=PLWHFCZSI5GFFUMN'
r = requests.get(url)
apple_macd_weekly = r.json()

print(apple_macd_weekly)
```

```python
apple_macd_weekly_df = pd.DataFrame.from_dict(apple_macd_weekly['Technical Analysis: MACD']).T
apple_macd_weekly_df.head(50)
```
