```python
import requests
import pandas as pd
```

```python
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY&symbol=GOOG&apikey=PLWHFCZSI5GFFUMN'
r = requests.get(url)
google_weekly_data = r.json()

print(google_weekly_data)
```

```python
google_weekly_df = pd.DataFrame.from_dict(google_weekly_data['Weekly Time Series']).T
google_weekly_df.head(50)
```

```python
url = 'https://www.alphavantage.co/query?function=RSI&symbol=GOOG&interval=weekly&time_period=10&series_type=open&apikey=PLWHFCZSI5GFFUMN'
r = requests.get(url)
google_rsi_weekly = r.json()

print(google_rsi_weekly)
```

```python
google_rsi_weekly_df = pd.DataFrame.from_dict(google_rsi_weekly['Technical Analysis: RSI']).T
google_rsi_weekly_df.head(50)
```

```python
google_weekly_df = google_weekly_df.reset_index()
```

```python
google_rsi_weekly_df = google_rsi_weekly_df.reset_index()
```

```python
merged_weekly_rsi = pd.merge(google_weekly_df, google_rsi_weekly_df , on=['index', 'index'])
merged_weekly_rsi.head(50)
```

```python
merged_weekly_rsi = merged_weekly_rsi.rename(columns={'index': 'Date', '1. open': 'Open', '2. high': 'High', '3. low': 'Low', '4. close': 'Close', '5. volume': 'Volume'})

merged_weekly_rsi
```

```python
merged_weekly_rsi.dropna()
```

```python
merged_weekly_rsi_adjusted = merged_weekly_rsi[['Date', 'Close', 'RSI']]
merged_weekly_rsi_adjusted
```

```python
merged_weekly_rsi_adjusted = merged_weekly_rsi_adjusted.sort_values('Date', ascending = True)
merged_weekly_rsi_adjusted.head(50)
```

```python
merged_weekly_rsi_adjusted.dtypes
```

```python
merged_weekly_rsi_adjusted['Close'] = pd.to_numeric(merged_weekly_rsi_adjusted['Close'], errors='coerce')
```

```python
merged_weekly_rsi_adjusted['RSI'] = pd.to_numeric(merged_weekly_rsi_adjusted['RSI'], errors='coerce')
```

```python
merged_weekly_rsi_adjusted.dtypes
```

```python
merged_weekly_rsi_adjusted['Buy'] = (merged_weekly_rsi_adjusted['RSI'] < 30) & (merged_weekly_rsi_adjusted['RSI'].shift(periods = 1) > 30) 
```

```python
merged_weekly_rsi_adjusted['Sell'] = (merged_weekly_rsi_adjusted['RSI'] > 70) & (merged_weekly_rsi_adjusted['RSI'].shift(periods = 1) < 70)
```

```python
merged_weekly_rsi_adjusted.head(50)
```

```python
df = merged_weekly_rsi_adjusted[((merged_weekly_rsi_adjusted['Sell'] == True) & (merged_weekly_rsi_adjusted['Sell'].shift(periods = 1) == False)) | ((merged_weekly_rsi_adjusted['Buy'] == True) & (merged_weekly_rsi_adjusted['Buy'].shift(periods = 1) == False))] 
df = df.reset_index()
df
```

```python
df2 = df[((df['Sell'] == True) & (df['Sell'].shift(periods = 1) == False)) | ((df['Buy'] == True) & (df['Buy'].shift(periods = 1) == False))] 
df2 = df2.reset_index()
df2
```

```python
import numpy as np
```

```python
cap = 10000
shares = 0
cap_list = []
share_list = []
df_test = df2.copy()

for i, row in df_test.iterrows():
    if row['Sell']:
        #print("sell")
        cap = shares * row['Close']
        shares = 0
        #print(f"1 cap = {cap} \t shares={shares}")

        cap_list.append(cap)
        share_list.append(0)
    else:
        #print("buy")
        shares = cap / row['Close']
        cap = 0
        #print(f"2 cap = {cap} \t shares={shares}")

        cap_list.append(0)
        share_list.append(shares)

    if cap < 0:
        print("BROKE!!!")
        break

if cap >= 0:
    df_test['capital'] = cap_list
    df_test['shares'] = share_list

df_test
```
